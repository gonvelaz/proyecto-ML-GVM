```python
import random
import pandas as pd

# Crear lista de ids con algunos que se repiten
ids = [random.randint(1, 10) for _ in range(40)]
ids[10:13] = [ids[0], ids[3], ids[6]]

# Crear diccionario aleatorio
data_dict = {
    'id': ids,
    'columna1': [random.random() for _ in range(40)],
    'columna2': [random.choice(['A', 'B', 'C']) for _ in range(40)],
    'columna3': [random.randint(1, 100) for _ in range(40)],
    'columna4': [random.uniform(0, 1) for _ in range(40)]
}
```

```python
df = pd.DataFrame(data_dict)
```

```python
df.head()
```

```python
# Ordenar el dataframe por fecha_timestamp
#df = df.sort_values(by='fecha_timestamp')

if df.groupby('id')['columna4'].shift(1).isna().all():
    
    df['goles_local_3_previos'] = df.groupby('id')['columna4'].shift(-1)
    df['tiros_local_3_previos'] = df.groupby('id')['columna3'].shift(-1)

elif df.groupby('id')['columna4'].shift(2).isna().all():
    df['goles_local_3_previos'] = df.groupby('id')['columna4'].shift(1)
    df['tiros_local_3_previos'] = df.groupby('id')['columna3'].shift(1)
    
elif df.groupby('id')['columna4'].shift(3).isna().all():
    # Obtener los goles de los 3 partidos anteriores
    df['goles_local_3_previos'] = df.groupby('id')['columna4'].shift(1) + \
                                  df.groupby('id')['columna4'].shift(2) 
    # Obtener los tiros de los 3 partidos anteriores
    df['tiros_local_3_previos'] = df.groupby('id')['columna3'].shift(1) + \
                                  df.groupby('id')['columna3'].shift(2) 

else:    
    # Obtener los goles de los 3 partidos anteriores
    df['goles_local_3_previos'] = df.groupby('id')['columna4'].shift(1) + \
                                  df.groupby('id')['columna4'].shift(2) + \
                                  df.groupby('id')['columna4'].shift(3)
    # Obtener los tiros de los 3 partidos anteriores
    df['tiros_local_3_previos'] = df.groupby('id')['columna3'].shift(1) + \
                                  df.groupby('id')['columna3'].shift(2) + \
                                  df.groupby('id')['columna3'].shift(3)
```

```python
# Hay suficientes valores anteriores para calcular los tres partidos previos
df['goles_local_3_previos'] = df.groupby('id')['columna4'].shift(1) + \
                              df.groupby('id')['columna4'].shift(2) + \
                              df.groupby('id')['columna4'].shift(3)
df['tiros_local_3_previos'] = df.groupby('id')['columna3'].shift(1) + \
                              df.groupby('id')['columna3'].shift(2) + \
                              df.groupby('id')['columna3'].shift(3)

df['goles_local_3_previos'] = df['goles_local_3_previos'].fillna(df.groupby('id')['columna4'].shift(-1))
df['tiros_local_3_previos'] = df['tiros_local_3_previos'].fillna(df.groupby('id')['columna3'].shift(-1))
```
